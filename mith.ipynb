```python
train_data = train_data.drop(['CustomerID','Location.Geo'],axis=1)
```

```python
train_data.columns
```

```python
catcols = ['Education','EmploymentStatus','Gender','Location.Code','Marital.Status','Policy.Type','Policy','Renew.Offer.Type','Sales.Channel','Vehicle.Class','Vehicle.Size']
```

```python
numcols = ['Income']
```

```python
train_data.Income=train_data.Income.replace('?',np.nan)
```

```python
train_data[catcols] = train_data[catcols].apply(lambda x: x.astype('category'))
train_data[numcols] = train_data[numcols].apply(lambda x: x.astype('float')) 
```

```python
train_data.dtypes
```

```python
train_data['Coverage'].fillna(train_data['Coverage'].mode()[0],inplace=True)
train_data['EmploymentStatus'].fillna(train_data['EmploymentStatus'].mode()[0],inplace=True)
train_data['Location.Code'].fillna(train_data['Location.Code'].mode()[0],inplace=True)
train_data['Marital.Status'].fillna(train_data['Marital.Status'].mode()[0],inplace=True)
train_data['Policy.Type'].fillna(train_data['Policy.Type'].mode()[0],inplace=True)
train_data['Vehicle.Class'].fillna(train_data['Vehicle.Class'].mode()[0],inplace=True)
train_data['Education'].fillna(train_data['Education'].mode()[0],inplace=True)
train_data['Gender'].fillna(train_data['Gender'].mode()[0],inplace=True)
train_data['Policy'].fillna(train_data['Policy'].mode()[0],inplace=True)
train_data['Renew.Offer.Type'].fillna(train_data['Renew.Offer.Type'].mode()[0],inplace=True)
train_data['Sales.Channel'].fillna(train_data['Sales.Channel'].mode()[0],inplace=True)
```

```python
train_data.isnull().sum()
```

```python
 missing_col = ['Number.of.Open.Complaints','Income','Monthly.Premium.Auto','Number.of.Policies','Vehicle.Size']
for i in missing_col:
  train_data.loc[train_data.loc[:,i].isnull(),i]=train_data.loc[:,i].mean()
```
